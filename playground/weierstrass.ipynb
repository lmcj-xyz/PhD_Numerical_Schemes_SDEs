```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
sys.path.insert(0, os.path.abspath(os.path.join(os.path.dirname('..'))))
import dsdes as ds
```

```python
rng = np.random.default_rng()
gaussian = rng.standard_normal(2**12)
f1 = ds.fbm(gaussian, 0.9, 2**12, 10)
f2 = ds.fbm(gaussian, 0.8, 2**12, 10)
f3 = ds.fbm(gaussian, 0.7, 2**12, 10)
```

Weierstrass function parameters

```python
alpha1, alpha2, alpha3 = 0.9, 0.6, 0.7
```

```python
x1, w1 = ds.weierstrass(alpha=alpha1, b=11)
x1_10, w1_10 = ds.weierstrass(alpha=alpha1, b=11, rescale=10)
x1_100, w1_100 = ds.weierstrass(alpha=alpha1, b=11, rescale=100)
```

```python
x2, w2 = ds.weierstrass(alpha2, 11)
x2_10, w2_10 = ds.weierstrass(alpha=alpha2, b=11, rescale=10)
x2_100, w2_100 = ds.weierstrass(alpha=alpha2, b=11, rescale=100)
```

```python
x3, w3 = ds.weierstrass(alpha3, 10)
```

```python
grid0 = np.linspace(0, 20, 2**12)
b1 = ds.bridge(f1, grid0)
b2 = ds.bridge(f2, grid0)
b3 = ds.bridge(f3, grid0)
```

```python
wd1, w1, x1, var1 = ds.wdrift(alpha=0.9, b=13, points=2**12, half_support=10, time_steps= 2**24)
wd2, w2, x2, var2 = ds.wdrift(alpha=0.6, b=13, points=2**12, half_support=10, time_steps= 2**26)
```

```python
plt.plot(x1, wd1, label=f'der alpha {alpha1}')
plt.plot(x2, wd2, label=f'der alpha {alpha2}')
plt.plot(x1, w1,  label=f'w alpha   {alpha1}')
plt.plot(x2, w2,  label=f'w alpha   {alpha2}')
plt.ylim(-10, 10)
plt.xlim(-1.5, 1.5)
plt.legend()
plt.show()
```

```python
plt.plot(x3, w3, label='alpha 0.7')
plt.plot(x1, b1, label='hurst 0.9')
plt.plot(x2, b2, label='hurst 0.8')
plt.plot(x3, b3, label='hurst 0.7')
plt.legend()
plt.show()
```
